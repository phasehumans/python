```python
data_customers['id']= data_customers['id'].astype(object)
data_purchases['id']= data_purchases['id'].astype(object)
data_products['id']= data_products['id'].astype(object)
data_customers['postcode']= data_customers['postcode'].astype(object)
data_customers['street_num']= data_customers['street_num'].astype(object)
```

```python
data_products.head(10)
```

```python
avg_company= data_products.groupby('company').agg({'cost': 'mean'})
avg_company.head(10)
```

```python
avg_cost_per_company= data_products.groupby('company')['cost'].mean().reset_index()

# reset_index() -- > convert this into dataframe (index 0,1,2 ..)
avg_cost_per_company.head(10)
```

```python
type(avg_cost_per_company)
```

```python
# data visulaization

plt.figure(figsize= (18,12))
plt.bar(avg_cost_per_company['company'], avg_cost_per_company['cost'])

plt.title('Average cost per company')
plt.xlabel('Company', fontsize= 12)
plt.ylabel('Average Cost', fontsize= 12)

plt.xticks(rotation= 90)
plt.grid(axis= 'y', linestyle= '--', linewidth=0.9)


# save graph
plt.savefig('average_product_cost_per_company.png')

plt.show()
```

Bar plot -> x axis (categorical data : represets groups / categerious (not numbers)
use for comparision

Box plot -> to detect outliers (outliners: points significant diff from others / extreme values)

```python
plt.figure(figsize=(10,8))

plt.boxplot(avg_cost_per_company['cost'])
plt.title('Box plot of cost of the product', fontsize= 16)
plt.ylabel('cost of the product')

plt.savefig('boxplot.png')
plt.show()

# here no outliers
```

## Seaborn
- seaborn built on top of matplotlib
- other options excel, powerbi, tablue etc
- no customization


```python
tips= sns.load_dataset('tips')
# tips inbuiilt dataset in seaborn
tips.head(10)
```

```python
tips
```

```python
# total tips per day

sns.barplot(x= 'day', y='tip', data= tips, estimator= sum)
# estimator is sum,mean, median etc

plt.title('Total tips by day')
plt.xlabel('day')
plt.ylabel('total tips')

plt.savefig('total_tips_per_day.png')
plt.show()

# lines -- confidence interval
```

```python
sns.kdeplot(data=tips, x='total_bill', hue= 'time')  #hue -> segregate
plt.title('KDE plot of the total bill')

plt.xlabel('total bill')
plt.show()

# KDE : visulaize probability density fn
```

```python
sns.boxplot(x= 'day', y= 'tip',data= tips)
plt.title('boxplot of tips')

plt.xlabel('day')
plt.ylabel('tips')

plt.savefig('outlier_tips.png')
plt.show()
```

```python
# Vionlinplot
sns.violinplot(x= 'day', y= 'tip',data= tips)
plt.title('vionlinplot of tips')

plt.xlabel('day')
plt.ylabel('tips')

plt.savefig('outlier_tips.png')
plt.show()
```

```python
sns.pairplot(data= tips)
```

```python
sns.regplot(x= 'total_bill', y='tip' ,data= tips)

# regression plot: best fit line, max density point collect
```

```python
sns.scatterplot(x='total_bill', y='tip', data= tips)
```

```python
sns.catplot(x='smoker', y= 'tip',data= tips)
```

```python
# for feature selection (correlation coefficent)

numeric_tips= tips.select_dtypes(include='number')
correlation_coefficent= numeric_tips.corr()
print(correlation_coefficent)
```

```python
sns.heatmap(correlation_coefficent, annot= True)
plt.title('Correlation Coefficent Matrix ')
plt.savefig('correlation.png')
plt.show()
```
