```python
# drop missing values

# create copy of data first
copy_data= data.copy()
copy_data
```

```python
copy_data.isnull().sum()
```

```python
# strategy 1: dropna()
copy_data= copy_data.dropna()
```

```python
copy_data.shape

# lost of data from 1000 to 418

# dropna() --> use when small portion of missing of data
```

```python
# strategy 2: data imputation

# imputation -> with the measures of central tendency[median, mean, mode]
# mode use when catogrical data ex. M/F
```

```python
mode_gender= data['gender'].mode()[0]

mode_gender

# when data is 50-50 then mode is male and female both, so use [0]

# first cater the imputation(missing) and then balance the data
```

```python
data['gender'].fillna(mode_gender, inplace=True)

data.isnull().sum()
```

```python
data.shape
```

```python
# numerical data --> mean, median, ffill, bfill

# ffill --> forward fill (above value is copied)
data['street_num']= data['street_num'].fillna(method= 'ffill')

data.isnull().sum()
```

```python
data['street_name']= data['street_name'].fillna(method= 'ffill')
data['street_suffix'].fillna(method= 'ffill', inplace= True)

data.isnull().sum()
```

```python
# bfill : backward fill (below cell)

data['city'].fillna(method= 'bfill', inplace= True)
data['state'].fillna(method= 'bfill', inplace= True)
data['postcode'].fillna(method= 'bfill', inplace= True)
```

```python
data.isnull().sum()
```

```python
# email unique, no sense to use ffill , bfill, so use parameter (x random)

data['email']= data['email'].fillna('unknown')
```

```python
data.isnull().sum()
```

```python
data_product['company']= data_product['company'].fillna('Unknown')

data_product.isnull().sum()
```
